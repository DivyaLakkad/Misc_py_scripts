```python
def d_col(row):
    date_str = row["Date"]
    try:
        date_object = datetime.strptime(date_str, '%A %B %d,%Y')
        new_date_format = datetime.strftime(date_object, '%d.%m.%Y')
    except:
        # try:
        date_object = datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')
        new_date_format = datetime.strftime(date_object, '%d.%m.%Y')
        # except:
            # date_object = datetime.strptime(date_str, '%Y%m%d')
            # new_date_format = datetime.strftime(date_object, '%d.%m.%Y')
    
    row["Date"] = new_date_format

    return row["Date"]
```

```python
df_with_dates['Date'] = df_with_dates.apply(lambda row: d_col (row), axis=1)
```
